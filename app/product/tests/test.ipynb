```
import requests

def call_insert_product_sync():
    url = "http://localhost:8001/product/insert"
    payload = {
        "product_id": 3,
        "product_name": "KALEM",
        "updated_at": "20170908"
    }

    response = requests.post(url, json=payload)
    print(response.json())

call_insert_product_sync()
```
