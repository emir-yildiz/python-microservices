```
import requests

def call_insert_product_sync():
    url = "http://localhost:8001/product/insert"
    payload = {
        "product_id": 4,
        "product_name": "DEFTER",
        "updated_at": "20180908"
    }

    response = requests.post(url, json=payload)
    print(response.json())

call_insert_product_sync()
```

```

def call_delete_product_sync():
    url = "http://localhost:8001/product/delete"
    payload = {"product_id": 4}  # ← dict olmalı

    response = requests.delete(url, json=payload)  # ← json= ile gönderilmeli
    print(response.json())

call_delete_product_sync()
```
